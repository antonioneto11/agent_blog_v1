## To create the agents and tasks for writing blog content on a given topic, we can follow these steps:

## Plan

Define Agents: Create four agents with specific roles: Planner, Researcher, Writer, and Editor.

Create a Custom Tool: Implement a tool using the SerperDev API for searching the internet.

Define Tasks: Establish tasks for planning, researching, writing, and editing.

Setup Crew: Assemble the agents and tasks into a crew.

1. Define Agents
Each agent will have a specific role and goal. The agents will use GPT-4 for generating content and the custom search tool for internet research

```python
# Warning control
import warnings
warnings.filterwarnings('ignore')
```

```python
import os
import requests
from crewai import Agent, Task, Crew, Process
from crewai_tools import BaseTool
from crewai_tools import SerperDevTool


from dotenv import load_dotenv
load_dotenv()
# Set API keys


serper_api_key = os.getenv('SERPER_API_KEY')
openai_api_key = os.getenv('OPENAI_API_KEY')

# Initialize the SerperDevTool for internet research
search_tool = SerperDevTool()
```

## Define Agents

With the custom tool defined, we can now create the agents using this tool.

```python
# Define agents
planner = Agent(
    role='Planner',
    goal='Plan how to write about the given {topic}',
    verbose=True,
    memory=True,
    backstory="You're a seasoned planner, skilled at outlining and structuring blog content effectively.",
    tools=[],
    allow_delegation=True
)

researcher = Agent(
    role='Researcher',
    goal='Research the {topic} on the internet',
    verbose=True,
    memory=True,
    backstory="You excel at finding and compiling information from various sources on the internet.",
    tools=[search_tool],
    allow_delegation=True
)

writer = Agent(
    role='Writer',
    goal='Write about the {topic} using the information provided by the researcher',
    verbose=True,
    memory=True,
    backstory="You are a creative writer, capable of crafting engaging and informative blog content.",
    tools=[],
    allow_delegation=False
)

editor = Agent(
    role='Editor',
    goal='Edit the content created by the writer',
    verbose=True,
    memory=True,
    backstory="As an editor, you have a keen eye for detail, ensuring the content is polished and error-free.",
    tools=[],
    allow_delegation=False
)
```

## Define Tasks

We set up the tasks for planning, researching, writing, and editing.

```python
# Define tasks
plan_task = Task(
    description=(
        "Create a detailed plan for the blog content on the  {topic}. "
        "Outline the structure, key points, and overall flow. "
        "Additionally, strategize alternative ways to gather information if the primary research tool encounters issues."
    ),
    expected_output='A structured plan detailing the sections, key points for the blog content, and alternative information collection strategies.',
    tools=[],
    agent=planner
)

research_task = Task(
    description="Research the {topic} on the internet and compile relevant information.",
    expected_output='A comprehensive research document with relevant information about the topic.',
    tools=[search_tool],
    agent=researcher
)

write_task = Task(
    description="Write a detailed blog post using the research provided by the researcher.",
    expected_output='A complete blog post draft about the topic.',
    tools=[],
    agent=writer
)

edit_task = Task(
    description="Edit the blog post draft for clarity, coherence, and correctness.",
    expected_output='A polished and final version of the blog post.',
    tools=[],
    agent=editor
)
```

### Setup Crew
 we assemble the crew and kick off the process.

```python
# Assemble the crew
crew = Crew(
    agents=[planner, researcher, writer, editor],
    tasks=[plan_task, research_task, write_task, edit_task],
    process=Process.sequential
)

# Kickoff the crew with a given topic
result = crew.kickoff(inputs={'topic': 'The Impact of AI on Healthcare'})
print(result)
```
